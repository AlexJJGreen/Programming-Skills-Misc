WHAT IS THE COMMAND LINE?

* The command line is an integral tool for making the jump from beginners programming to a more compentent and efficient level. Essentially it is a way to skip the use of GUIs (graphical user interfaces) and talk to the computer directly using text commands (essentially this is what is going on in the background when you click on things in the GUI). A great example of this is navigating to a folder on your desktop. In a graphical interface, the desktop is usually the first screen you will see after the login. With the mouse you then hover over the file you want and double click the mouse or keypad. In the command line you can do exactly the same thing just by typing a command like this:

cd desktop/*name of folder*

 - This means "change directory to the desktop and then change again to the folder
 
* In this example, the time saved typing instead of clicking is minimal, but for larger more complex tasks, the command line can be an essential tool to help speed up the process.

* The command line is also used to interact with programs and programming languages. Python scripts are normally run directly in the command line window. The command line also allows you quickly to download thing you may need to help with coding such as Python Packages or tools to handle tasks that can be automated in the background.

USING THE COMMAND LINE

To start the windows cmd line tool:
    * Type cmd into your windows search bar and the option should come up. SImply click on it and it should start running.
    
You will see when you open it, a blinky dot next to some writing that looks like

C:\Users\JohnSmith>

This is your current working directory. That means this is the directory of files and folders (other directories) that you are currently in. 

To change directory you type in the command

cd *path of directory*

Now path is not the same as name. The computer must know the exact path of the directory you want to change to from the current directory you're in. So for example if you are in your user directory, your desktop should be a directly available from it ie:

cd desktop

but if you hada file of pictures on your desktop called "pics" that you wanted to go to, you would have to write

cd desktop/pics

To go back to the desktop from your pics directory use:

cd ..

The .. notation means "go to the parent directory" or one directory up. This can be chained together from you pics file to go straight back to your users directory in the same way as you chained the dekstop and pics directories, by using forwars slash, ie:

cd ../..

if you are supe efficient, you can use tab to autcomplete directories for you.

LISTING CONTENTS OF A DIRECTORY

To show a list of the contents of a directory, aka. the files and other directories inside it, use the command:

dir

This will list a directories contents. It will also show you the size of any files and the total size of all directories combined

To list the contents of another directory simply use the same path technique for changing directory:

dir *path of directory*

To see all the 'hidden directories' use:

dir /a

BUT! you probably shouldn't mess with hidden files

To list just a certain type of file use:

dir * *file suffix aka. .jpg*

To go back through your command history (in case you don't want to type a certain command again) you can use the up and down arrow keys to cycle through your previous commands.

OPEN A FILE

To open a file, navigatate to its directory (or know its full path from your current directory) and simply type in the full name with its file extension and this will run the file in its default program/app


CLEAR THE SCREEN

The cmd line can get very cluttered so to clear the screen, type:

cls

HELP COMMAND

If you want to know about any command you can type the command and then:

/?

And you will receive a list of all the things you can do with it

CREATING AND DELETING DIRECTORIES

To create a directory type

mkdir *name of directory*

to remove a directory

rmdir *name of directory*

but! if anything is in the directory, it won't remove it so, to remove a directory and all its contents, type

rmdir /s *name of directory*

PATH

For windows to run executable files (programs basically) its directory needs to be in the windows system path. Type:

path

and you will see all the different directories that the computer will scan through to run an exectuable. You can add directories to the path in settings/environment variables.

DRIVES

A drive is a part of the computer that you can read data from and write data to (a chunk of memory). THe most common is the local disk or C: Drive. THe harddrive is C: because A and B where reserved for floppy disk drives back in the day.

Typing:

wmic logicaldisk get name

will give you a list of drives.

To switch drives type in the letter of the drive followed by : ie:

E:

DELETING AND APPENDING TO FILES

To delete a file type:

del *file name*

To read a file:

type *file name*

To write a file:

echo *what you want to write* > *name of file*

To append to a file:

echo *what you want to write >> *name of file*

COPYING AND MOVING FILES

To copy a file:
    
copy *name of file* *name of folder to move it to*

To copy a directory and its subdirectories:

xcopy *name of dir to copy* *dir to copy to* /s

To move it!

move *name of file/ directory* *name of new directory*

RENAMING A DIR

rename *name of file* *files new name*
